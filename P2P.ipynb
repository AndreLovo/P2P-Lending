```python
# Importando as bibliotecas
import pandas as pd
import requests

# importando uma nova biblioteca para gráficos
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Configurando o matplotlib
#%matplotlib inline

#gráfico interativo como figura
%matplotlib notebook 
from matplotlib import cm

# Montando o gráfico 3d
# mpl => matplotlib
from mpl_toolkits.mplot3d import Axes3D
```

```python
# URL
url='https://valorinveste.globo.com/produtos/credito/noticia/2021/11/02/emprestimo-entre-pessoas-p2p-pode-garantir-rendimento-de-dois-digitos-mas-risco-e-alto.ghtml'

# Requisitar a página e guardar a resposta
resposta = requests.get(url)

#Obter a resposta em formato HTML
table = pd.read_html(resposta.text)

#Vamos dar uma olhada no resultado
table
```

```python
df = table[0]
display(df)
```

```python
#subindo a linha 0
df.columns = df.iloc[0]
#tirando a linha 0
df=df[1:7]
display(df)
```

```python
# Verificando as informações da tabela 
df.info()
```

```python
list(df["RETORNO MIN. E MÁX. AO ANO"])
```

```python
# Substituindo os valores
# https://www.delftstack.com/pt/howto/python-pandas/pandas-replace-values-in-column/
#df[column_name].replace([old_value1, old_value2, old_value3],[new_value1, new_value2, new_value3])
df["RETORNO MIN. E MÁX. AO ANO"] = df["RETORNO MIN. E MÁX. AO ANO"].replace(['18% a 36%','12% a 42%','14% a 42%','12% a 60%','31,37% a 56,08%','12% a 40%'],
                                                                            ['18 a 36', '12 a 42', '14 a 42', '12 a 60', '31.37 a 56.08', '12 a 40'])
df["RETORNO MIN. E MÁX. AO ANO"]
```

```python
#Separando os valores dos retornos Mínimos e Máximos
# As informações estão separadas pela letra a
#df[[0, 1]] = tabela50f2["Jogos"].str.split(" ", expand=True)

df[[0, 1]] =df["RETORNO MIN. E MÁX. AO ANO"].str.split(" a ", expand=True)
display(df)
```

```python
#Retirando a coluna que foi separada
df=df.drop("RETORNO MIN. E MÁX. AO ANO", axis=1)
display(df)
```

```python
# Alterando os tipos dos dados para string
df[0] = df[0].astype('float64')
df[1] = df[1].astype('float64')
```

```python
df.info()
```

```python
#Renomeando as colunas 0 e 1 em 'RETORNO MIN. AO ANO', 'RETORNO MÁX. AO ANO'
df.rename(columns={0: 'RETORNO MIN. AO ANO'}, inplace=True)
df.rename(columns={1: 'RETORNO MÁX. AO ANO'}, inplace=True)
#Renomeando as outras colunas
df.rename(columns={1: ' %'}, inplace=True)
```

```python
df.info()
```

```python
# Substituindo os valores
# https://www.delftstack.com/pt/howto/python-pandas/pandas-replace-values-in-column/
#df[column_name].replace([old_value1, old_value2, old_value3],[new_value1, new_value2, new_value3])
```

```python
list(df["RETORNO MÉDIO AO ANO"])
```

```python
# Substituindo os valores
# https://www.delftstack.com/pt/howto/python-pandas/pandas-replace-values-in-column/
#df[column_name].replace([old_value1, old_value2, old_value3],[new_value1, new_value2, new_value3])
df["RETORNO MÉDIO AO ANO"]=df["RETORNO MÉDIO AO ANO"].replace(['25%', '27%', '14,3%', '21,27%', '25,3%', '22%'],[25, 27, 14.3, 21.27, 25.3, 22])
df["RETORNO MÉDIO AO ANO"]
```

```python
# Substituindo os valores
# https://www.delftstack.com/pt/howto/python-pandas/pandas-replace-values-in-column/
#df[column_name].replace([old_value1, old_value2, old_value3],[new_value1, new_value2, new_value3])
```

```python
list(df["INADIMPLÊNCIA"])
```

```python
# Substituindo os valores
# https://www.delftstack.com/pt/howto/python-pandas/pandas-replace-values-in-column/
#df[column_name].replace([old_value1, old_value2, old_value3],[new_value1, new_value2, new_value3])
df["INADIMPLÊNCIA"]=df["INADIMPLÊNCIA"].replace(['9%', '3,7%', '8,1%', '5,53%', '8%', '5,4%'],[9, 3.7, 8.1, 5.53, 8, 5.4])
df["INADIMPLÊNCIA"]
```

```python
list(df["INVESTIMENTO MÍNIMO"])
```

```python
# Substituindo os valores
# https://www.delftstack.com/pt/howto/python-pandas/pandas-replace-values-in-column/
#df[column_name].replace([old_value1, old_value2, old_value3],[new_value1, new_value2, new_value3])
#Obs: Fizemos esta etapa para verificarmos as diferentes formas de substituição de valores
df["INVESTIMENTO MÍNIMO"]=df["INVESTIMENTO MÍNIMO"].replace(['R$ 1.000,00','R$ 2.000,00','R$ 1.000,00','R$ 500,00','R$ 500,00','R$ 2.000,00'],['1.000,00','2000.00','1000.00','500.00','500.00','2000.00'])
```
